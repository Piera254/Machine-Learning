```python
#load wine dataset in scikit learn is what we use.
import pandas as pd
from sklearn.datasets import load_wine
def load_wine_data():
    wine_data =  load_wine()
    df = pd.DataFrame (data = wine_data.data, columns = wine_data.feature_names)
    df['target'] = wine_data.target#the adding the target(wine quality clas) 
    return df
```

```python
#model_training.py
if __name__ == "__main__":
    df = load_wine_data()
    print(df.head())
    print(df.describe())
    print(df["target"].value_counts())#distribution of wine classes
```

```python
#splitting the data into training and test sets  and scale the features

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

def preprocess_data(df):

    X = df.drop('target', axis = 1)#features
    y = df['target']#target wine quality

    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state =  27)
    #feature scaling
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)
    
    return X_train_scaled, X_test_scaled, y_train, y_test
#the standardscaler ensures all features contribute equally to the model
```

```python
#Training the logistic regression model and saving it in a pickle file
from sklearn.linear_model import LogisticRegression
import pickle

def train_model(X_train, y_train):
    model = LogisticRegression(random_state = 42)
    model.fit(X_train, y_train)
    #save the trained model using pickle
    with open('classifier.pkl', 'wb') as f:
        pickle.dump(model, f)
    return model
        
    
```

```python
#Evaluating the model
from sklearn.metrics import accuracy_score

def evaluate_model(model, X_test, y_test):
    y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    print(f"Accuracy: {accuracy:2f}")

if __name__ == "__main__" :
    df = load_wine_data()
    X_train_scaled, X_test_scaled, y_train, y_test = preprocess_data (df)
    model = train_model(X_train_scaled, y_train)
    evaluate_model(model, X_test_scaled, y_test)
    
    
```

# model validation techniques #
12 must know methods to validate my machin learning
1. Validation methods test machine predictions to measure their reliability

2. Model validation- process of testing how well a machine learning model works with data it has not seen or used in training.
   Existing data is used to check a model's performance helping in indentifying problems before deploying the model for real use.

4. Validation methods can produce different results thus choosing the right method matters.

5. Some validation techniques work better with specific data types and models 

6. Incorrect validation techniques can give misleading results about the model's true performance.

7.  There are two types of model validation techniques. hold-out methods and cross-validation methods.

8. hold-out methods includes the Train-Test-Split  --- Train-Validation-Test Split

9. CROSS VALIDATION METHODS ARE GROUPED INTO K-fold, Leav-out and Random methods.

10. K-Fold Methods include the basic K-Fold method and Time Series Split. Basic K-Fold is then grouped into Stratified, Repeated, and Group K-Fold.
    Repeated K-Fold is further branched to Repeated Stratified K-Fold.

11. Leave out methods include Leave P-out which braches to Leave-one out.

12. Random Method is divided into shuffle-Split which braches to Stratified Shuffle-Split.

# Train-validation -test-split #

1. Here the data is split into 3 ...The middle part is called the validation data and is used to tune the parameters of the model and we are aiming to
   have the least amount of error here

2. Since the validation results is considered many times during this tuning process, our model might start doing too well on the validation as desired.

3. This is the reason why we make the separate test set. We only test it once at the very end...this gives the truth of how the model works

4. Typical ways of splitting data.Small data sets--60: 20: 20 ratio, Medium(70: 15 :15), and Large (80: 10: 10).

## Stratified k-fold ##

```python
#WOORKS BY MAKING SURE EACH SPLIT HAS THE SAME MIX AS THE ORIGInal data makin testing reliable unlike the basic kfold
from sklearn.model_selection import StratifiedKFold, cross_val_score
#cross validation strategy
cv = StratifiedKFold (n_splits = 3, shuffle = True, random_state = 42)

#calculate the cross validation score
scores = cross_val_score(dt, X_train, y_train, cv=cv)
print(f"Validation accuracy: {scores.mean():.3f} \pm {scores.std():.3f}")

#plot the trees for each split
plt.figure(figsize = (5, 4*cv.get_n_splits(X_train)))
for i, (train_idx, val_idx) in enumerate(cv.split (X_train, y_train)):
    #Train and visualize the treefor this split
    dt.fit(X_train.iloc[train_idx], y_train.iloc[train_idx])
    plt.subplot(cv.get_n_splits(X_train), 1, i+1)
    plot_tree(dt, feature_names = X_train.columns, impurity = False, filled = True, rounded = True)
    plt.title(f'split{i + 1} (Validation Accuracy: {scores[i]:.3f})\nTrain indices: {train_idx}\nValidation indices: {val_idx}')
plt.tight_layout()

```

## Repeated KFOLD ##

```python
#HERE THE K-FOLD PROCESS IS REPEATED MULTIPLE TIMES USING DIFFERENT RANDOM SPLITS EACH TIME.
#THIS Solves the problem of change of results btwn dofferent random splts
from sklearn.model_selection import RepeatedKFold

#cross validation strategy
n_splits = 3
cv = RepeatedKFold (n_splits = n_splits, n_repeats = 2, random_state = 42)

#calculate cross valiidation scores
scores = cross_val_score(dt, X_train, y_train, cv=cv)
print(f"Validation accuracy: {scores.mean():.3f} \pm {scores.std():.3f}")

#plot trees for each split
total_splits = cv.get_n_splits(X_train) #will be 6(3 folds * 2 repetitions)
plt.figure (figsize = (5, 4*total_splits))
for i, (train_idx, val_idx) in enumerate(cv.split(X_train, y_train)):
    #train and visualize the tree for this split
    dt.fit(X_train.iloc[train_idx], y_train.iloc[train_idx])
    #calculate repetition and fold numbers
    repetition, fold = i // n_splits+ 1, i % n_splits + 1
    plt.subplot(total_splits, 1, i + 1)
    plot_tree (dt, feature_names = X_train.columns, impurity = False, filled = True, rounded = True)
    plt.title(f'split {repetition}.{fold} (Validation Accuracy: {scores[i]:.3f})\n'
              f'Train indices: {list(train_idx)}\n'
              f'Validation indices: {list (val_idx)}')
plt.tight_layout()



   

```

## Repeated Stratified K-Fold ##
1. The method combines keeping class balance (stratification) and running multiple rounds of testing(repetition)

2.  It keeps the ridgt of diff mixes of of diff types of data while testing many times
   
3.  This works best for uneven small datasets

```python
from sklearn.model_selection import RepeatedStratifiedKFold

#cross validation strategy
n_splits = 3
cv = RepeatedStratifiedKFold(n_splits = n_splits, n_repeats = 2, random_state = 42)

#calculate cross-validation scores
scores = cross_val_score(dt, X_train, y_train, cv=cv)
print(f"Valifdation accuracy: {scores.mean():.3f}, {scores.std():.3f}")

#plot trees for each split
total_splits = cv.get_n_splits(X_train) #will be 6 (3 splits * 2 repetitions)
plt.figure(figsize = (5, 4*total_splits))
for i, (train_idx, val_idx) in enumerate (cv.split(X_train, y_train)):
    #Train and visualize the Tree for this split
    dt.fit(X_train.iloc[train_idx], y_train.iloc[train_idx])
    # calculate repetition and fold numberrs
    repetition, fold = i//n_splits + 1, i % n_splits + 1
    plt.subplot(total_splits, 1, i+1)
    plot_tree(dt, feature_names = X_train.columns, impurity = False, filled = True, rounded = True)
    plt.title(f'Split {repetition}.{fold} (Validation Accuracy: {scores[i]:.3f})\n'
               f'Train indices: {list(train_idx)}\n'
               f'Validation indices: {list(val_idx)}')
    
    


```

## Group  K-Fold ##
1. Sometimes our data Naturally comes in groups that should stay together.E.g the golf data where we have many mearsurements from the same golf course t
   throughout the year.
2. If we put some mearsurementtts from one golf course in training data and others in test data, we create a problem.:our model would indirectly learn
   learn about the test data during training because it saw other mearsurements from the same course.
3. Group k-fold fixes this by keeping mearsurements from the same group (like all mearsurements from one golf course)  together in the same part when
    we split the data.This prevents the model from seeing information it shouldn't, which could make us think it performs betterthan it really does.


```python
from sklearn.model_selection import GroupKFold
#create groups
groups = ['Group 1', 'Group 4','Group 5', 'Group 3', 'Group 1', 'Group 2', 'Group 4',
          'Group 2', 'Group 6', 'Group 3', 'Group 6', 'Group 5', 'Group 1', 'Group 4',
          'Group 4', 'Group 3', 'Group 1', 'Group 5', 'Group 6', 'Group 2', 'Group 4',
          'Group 5', 'Group 1', 'Group 4', 'Group 5', 'Group 5', 'Group 2', 'Group 6']

#simple Train-Test split
X_train, X_test, y_train, y_test, groups_train, groups_test = train_test_split(X, y, groups, test_size = 0.5, shuffle = False)

#cross validation strategy
cv = GroupKFold(n_splits = 3)

#calculate cross_validation scores
scores = cross_val_score (dt, X_train, y_train, cv = cv.split(X_train, y_train, groups = groups_train))
print(f"Validation accuracy: {scores.mean():.3f} \pm {scores.std():.3f}")

#plot trees for each split
plt.figure( figsize = (4, 3.5*cv.get_n_splits(X_train)))
for i, (train_idx, val_idx) in enumerate(cv.split(X_train, y_train, groups = groups_train)):
    #Get the groups for this split
    train_groups = sorted(set(np.array(groups_train)[train_idx]))
    val_groups = sorted(set(np.array(groups_train)[val_idx]))
    #Train and visualizee the tree for this split
    dt.fit(X_train.iloc[train_idx], y_train.iloc[train_idx])
    plt.subplot (cv.get_n_splits(X_train), 1, i+ 1)
    plot_tree(dt, feature_names = X_train.columns, impurity = False, filled = True, rounded = True)
    plt.title(f'split {1+1} (Validation Accuracy: {scores[i]:.3f})\n'
              f'Train indices: {train_idx} ({", ".join(train_groups)})\n'
              f'Validation indices: {val_idx} ({", ".join(val_groups)})')
plt.tight_layout()
    

```

## Time series split ##
1. In k-fold we randomly split the data with the assumption that one data piece does not affect the other
2. This however does not work with data that changes over time where what happened before affects what happens next.Time series split changes K-fold to    work  better with this kind of time ordered data.
3. Instead of splitting data randomly, time series split  uses data in order from past to future.The training data only includes information from times
   before the testing data. This matches how we use models in real life, where we use past data to predict what will happen next.
   

```python
from sklearn.model_selection import TimeSeriesSplit, cross_val_score

#cross_validation strategy
cv = TimeSeriesSplit(n_splits = 3)
#calculate the cross validation scores
scores = cross_val_score(dt, X_train, y_train, cv = cv)
print(f"Validation accuracy: {scores.mean():.3f} \pm {scores.std():.3f}")

#plot trees for each split
plt.figure(figsize = (4, 3.5*cv.get_n_splits(X_train)))
for i, (train_idx, val_idx) in enumerate (cv.split(X_train, y_train)):
    #train and visualize the tree
    dt.fit (X_train.iloc[train_idx], y_train.iloc[train_idx])
    plt.subplot(cv.get_n_splits(X_train), 1, i + 1)
    plot_tree(dt, feature_names = X_train.columns, impurity = False, filled = True, rounded = True)
    plt.title(f'Split {i+1} (Validation Accuracy: {scores[i]:.3f})\n'
              f'Train indices: {train_idx}\n'
              f'Validation indices: {val_idx}')
plt.tight_layout()
    
```

## Leave P-out cross validation ##
1. Works like leave-one-out but uses P pieces of data for testing instead of just one piece of data.Hence balancing btwn leave one out and kfold
2. The main problem here  is quick growth off the number of possible combinations

```python
from sklearn.model_selection import LeavePOut, cross_val_score

#Cross validation strategy
cv = LeavePOut(p = 3)

#calculate cross_validationscores (using all splits for accuracy)
scores = cross_val_score(dt, X_train, y_train, cv = cv)
print(f"Valudation accuracy: {scores.mean():.3f} \pmv{scores.std():3f}")

#plot the first 15 trees
n_trees = 15
plt.figure(figsize = (4, 3.5*n_trees))
for i, (train_idx, val_idx) in enumerate (cv.split(X_train, y_train)):
    if i >= n_trees:
        break
    
    #TRAIN AND VISUALIZE THE TREES
    dt.fit(X_train.iloc[train_idx], y_train.iloc[train_idx])
    plt.subplot(n_trees, 1, i + 1)
    plot_tree(dt, feature_names = X_train.columns, impurity = False, filled = True, rounded = True)
    plt.title(f'Split {i + 1} (Validation Accuracy: {scores[i]:.3f})\n'
              f'Train indices: {train_idx}\n'
              f'Validation indices: {val_idx}')

plt.tight_layout()

```

## Random Methods ##
1. shuffle Split.Uses completely random splits instead odf splitting data in an organized way.This method workd well especillyfor largedatsets where k-    fold may take long to run.

```python
from sklearn.model_selection import ShuffleSplit, train_test_split, cross_val_score

#cross_validation_strategy
cv = ShuffleSplit(n_splits = 3, test_size = 0.2, random_state = 41)

#calculate cross_validation scores
scores = cross_val_score (dt, X_train, y_train, cv = cv)
print(f"Validation accuracy: {scores.mean():.3f} \pm {scores.std():.3f}")

#plot tres for each split
plt.figure(figsize = (4, 3.5*cv.get_n_splits(X_train)))
for i, (train_idx, val_idx) in enumerate (cv.split(X_train, y_train)):
    #train and visualize
    dt.fit(X_train.iloc[train_idx], y_train.iloc[train_idx])
    plt.subplot(cv.get_n_splits(X_train), 1, i + 1)
    plot_tree(dt, feature_names = X_train.columns, impurity = False, filled = True, rounded = True)
    plt.title(f'Split {i+1} (Validation Accuracy: {scores[i]:.3f})\n'
               f'Train indices: {train_idx}\n'
               f'Validation indices: {val_idx}')


           

```

## Stratified shuffle split ##
1. Stratified shuffle split combines random splitting with keeping the right mix of diff types of data.Like Stratified K-Fold, it makes sure
    each split has about the same percentage of each type as the full data set.This wporks well for an uneven dataset

```python
from sklearn.model_selection import StratifiedShuffleSplit, train_test_split

#cross_validation Strategy
cv = StratifiedShuffleSplit(n_splits = 3, test_size = 0.2, random_state = 41)

#calculate cross_validation strategy
scores = cross_val_score(dt, X_train, y_train, cv = cv)
print(f"Validation accuracy: {scores.mean():.3f} \pm {scores.std():.3f}")

#plot trees for each split
plt.figure(figsize = (4, 3.5*cv.get_n_splits(X_train)))
for i, (train_idx, val_idx) in enumerate(cv.split(X_train, y_train)):
    
    dt.fit(X_train.iloc[train_idx], y_train.iloc[train_idx])
    plt.subplot(cv.get_n_splits(X_train), 1, i + 1)
    plot_tree(dt, feature_names = X_train.columns, impurity = False, filled = True, rounded = True)
    plt.title(f'Split {i +1} (Validation Accuracy: {scores[i]:.3f})\n'
              f"Train indices: {train_idx}\n"
              f"Validation indices: {val_idx}")
plt.tight_layout()
```
